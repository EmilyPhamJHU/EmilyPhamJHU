```python
%%R
population_info <- read.csv("county_pop_arcos.csv")
annual_info <- read.csv("county_annual.csv")
land_info <- read.csv("land_area.csv")

population_info_stripped <- subset(population_info, select = -c(countyfips, county_name, NAME, variable))

annual_info_update <- within(annual_info, countyfips[BUYER_STATE == 'AR' & BUYER_COUNTY == 'MONTGOMERY'] <- '05097')

subset(annual_info_update, countyfips == "05097") #check to see if countyfips has been replaced successfully
annual_info_update <- annual_info_update[annual_info_update$BUYER_COUNTY != 'NA', ]
subset(annual_info_update, BUYER_COUNTY == "NA") #check to see if BUYER_COUNTY has been replaced successfully

install.packages("dplyr") #had to install library to use select command
library(dplyr)

land_area <- select(land_info, Areaname, STCOU, LND110210D)
colnames(land_area)[2] <- "countyfips" #rename STCOU to countyfips

county_info <- merge(x=population_info,y=land_area,
             by="countyfips", all.x=TRUE) #left join full population info onto land_area
```
